```python
Q1. What is a database? Differentiate between SQL and NoSQL databases.
Ans :
Database :-A database is a structured collection of data that is organized in a way that allows for 
efficient storage, retrieval, and manipulation of information. Databases are used to store and manage 
data for a wide range of applications, from simple address books to complex enterprise systems. 

Difference between SQL and NoSQL :-

SQL Databases :
1.Structured Data: SQL databases are designed for structured data with well-defined schemas.
Data is organized into tables with rows and columns, and each column has a specific data type.

2.ACID Transactions: SQL databases adhere to the ACID (Atomicity, Consistency, Isolation, Durability) 
properties, which ensure data integrity and reliability. Transactions are either fully completed 
or fully rolled back in case of failures.

3.Relational: SQL databases are often referred to as relational databases because they support 
relationships between tables through foreign keys. This enables the creation of complex queries 
and joins to retrieve data.

4.Scalability: SQL databases typically follow a vertical scaling model, where you increase the
power of a single server (vertical scaling) to handle more data and traffic. This can be
cost-prohibitive for large-scale applications.

5.Examples: MySQL, PostgreSQL, Oracle, Microsoft SQL Server, SQLite.

NoSQL Databases :
    
1.Unstructured or Semi-structured Data: NoSQL databases are designed to handle a variety
of data types, including unstructured and semi-structured data. They do not enforce a
strict schema, allowing for more flexibility.

2.BASE Transactions: NoSQL databases often use the BASE (Basically Available, Soft state,
Eventually consistent) model, which sacrifices some of the ACID properties in favor of 
improved availability and scalability. Data consistency may be eventual rather than immediate. 

3.Non-Relational: NoSQL databases are non-relational, meaning they don't use tables with 
fixed schemas. They can use various data models, including document-oriented, key-value,
column-family, and graph databases.

4.Scalability: NoSQL databases typically follow a horizontal scaling model, where you add
more commodity servers to handle increasing data and traffic. This allows for better 
scalability and cost-effectiveness for large-scale applications.

5.Examples: MongoDB (document-oriented), Cassandra (column-family), Redis (key-value), 
Neo4j (graph), Amazon DynamoDB (key-value and document).                                                         
```

```python
Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.
Ans :
DDL:- DDL stands for Data Definition Language. It is a subset of SQL (Structured Query Language)
used to define, manage, and manipulate the structure of a database. DDL statements are used 
to create, modify, and delete database objects such as tables, indexes, and constraints. 
These statements do not deal with the actual data stored in the database but rather focus 
on the structure and organization of the data.

1.CREATE: The CREATE statement is used to create new database objects, such as tables, 
indexes, and views. It defines the structure of the object, including its columns, 
data types, constraints, and other properties.
Example :
CREATE TABLE employees (
    employee_id INT PRIMARY KEY,
    first_name VARCHAR(50),
    last_name VARCHAR(50),
    hire_date DATE
);

In this example, a new table named "employees" is created with columns for employee information.

2.DROP: The DROP statement is used to delete database objects, such as tables, indexes, or views, 
and all the data associated with them. This action is irreversible, and it permanently removes 
the object from the database.
Example :
DROP TABLE employees;

This statement will delete the "employees" table and all its data.

3.ALTER: The ALTER statement is used to modify the structure of an existing database 
object, such as adding, modifying, or dropping columns, constraints, or indexes.
Example :
ALTER TABLE employees
ADD COLUMN email VARCHAR(100);

This statement adds a new "email" column to the existing "employees" table.

4.TRUNCATE: The TRUNCATE statement is used to remove all data from a table while 
retaining the table structure. Unlike the DROP statement, TRUNCATE does not 
delete the table itself; it only removes the rows of data within it.
Example :
TRUNCATE TABLE employees;

This statement will remove all records from the "employees" table, but the table structure remains intact.
```

```python
Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.
Ans :
DML :-DML stands for Data Manipulation Language. It is a subset of SQL (Structured Query Language)
used for manipulating data stored in a database. DML statements allow you to perform operations
such as inserting, updating, and deleting data within database tables. Unlike 
DDL (Data Definition Language), which deals with the structure of the database, DML 
focuses on working with the actual data.

1.INSERT: The INSERT statement is used to add new rows of data into a table.
Example :
INSERT INTO employees (employee_id, first_name, last_name, hire_date)
VALUES (1, 'John', 'Doe', '2023-01-15');

In this example, a new row is inserted into the "employees" table, providing values for the specified columns.

2.UPDATE: The UPDATE statement is used to modify existing data within a table. It allows you to change values 
in one or more columns based on specified conditions.
Example :
UPDATE employees
SET first_name = 'Jane'
WHERE employee_id = 1;

This statement updates the "first_name" column in the "employees" table for the employee with 
an "employee_id" of 1, changing it from 'John' to 'Jane'.

3.DELETE: The DELETE statement is used to remove rows from a table based on specified conditions. 
If you don't specify any conditions, it will delete all rows in the table.
Example :
DELETE FROM employees
WHERE employee_id = 1;

This statement deletes the row with an "employee_id" of 1 from the "employees" table.
```

```python
Q4. What is DQL? Explain SELECT with an example.
Ans :-
DQL :- DQL stands for Data Query Language. It is a subset of SQL (Structured Query Language) used for 
querying and retrieving data from a database. DQL is primarily associated with the SELECT statement, 
which is used to retrieve data from one or more database tables based on specified criteria.

Here is an explanation of the SELECT statement with an example:
    
SELECT: The SELECT statement is used to retrieve data from one or more tables in a database. 
It allows you to specify the columns you want to retrieve and the conditions that must be met 
for the rows to be included in the result set. The basic syntax of a SELECT statement is as follows:
    
SELECT column1, column2, ...
FROM table_name
WHERE condition;

column1, column2, ...: These are the columns you want to retrieve from the table. We can use an asterisk * to select all columns.
table_name: This is the name of the table from which you want to retrieve data.
condition (optional): This is an optional part of the statement where you can specify criteria that must be met for rows 
to be included in the result set. If no condition is provided, all rows from the table will be returned.
```

```python
Q5. Explain Primary Key and Foreign Key.
Ans :-
Primary Key and Foreign Key are two important concepts in relational databases that define
relationships between tables and ensure data integrity. They serve distinct purposes:
    
Primary Key:

1.Uniqueness: A primary key is a column (or a set of columns) in a table that uniquely 
identifies each row in that table. It enforces the uniqueness constraint, meaning 
that no two rows in the table can have the same values in the primary key column(s).

2.Uniqueness and Not Null: In addition to uniqueness, a primary key column cannot 
contain NULL values. Every row must have a valid, non-null value in the primary key column(s).

3.Data Integrity: The primary key is used to establish data integrity and ensure that 
each row can be uniquely identified. It enables efficient data retrieval and indexing.

4.Example: In a "Students" table, a "student_id" column can be designated as the primary key. 
Each student will have a unique "student_id," and this key helps in identifying and 
referencing individual students.

Foreign Key:
    
1.Relationships: A foreign key is a column (or a set of columns) in one table that 
establishes a link between the data in two tables. It creates a relationship between 
the table containing the foreign key and another table (usually the table with the 
corresponding primary key).

2.Referential Integrity: The foreign key enforces referential integrity by ensuring 
that values in the foreign key column(s) match values in the primary key column(s) 
of the referenced table. This means that the data in the foreign key column(s) must 
exist in the referenced table.

3.Cascading Actions: You can specify actions that should occur when a record in 
the referenced table is modified or deleted. These actions include 
CASCADE (changes propagate to related rows), SET NULL (foreign key values set to NULL),
SET DEFAULT, and NO ACTION (no action allowed if related data exists).

4.Example: In a database with "Students" and "Courses" tables, the "student_id" column
in the "Courses" table can be a foreign key that references the "student_id" primary 
key in the "Students" table. This establishes a relationship between students and the 
courses they are enrolled in.
```

```python
Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.
Ans :
To connect MySQL to Python, you can use the mysql-connector-python library, which provides a 
Python interface for MySQL databases. 

pip install mysql-connector-python
Here's a Python code example that connects to a MySQL database, creates a cursor, and executes 
SQL queries using the cursor() and execute() methods:

import mysql.connector

# Replace with your MySQL database credentials
db_config = {
    "host": "your_host",
    "user": "your_username",
    "password": "your_password",
    "database": "your_database",
}

# Establish a connection to the MySQL database
connection = mysql.connector.connect(**db_config)

try:
    # Create a cursor object to interact with the database
    cursor = connection.cursor()

    # Execute SQL queries using the cursor

    # Example 1: Select data from a table
    cursor.execute("SELECT * FROM your_table")
    result = cursor.fetchall()
    for row in result:
        print(row)

    # Example 2: Insert data into a table
    insert_query = "INSERT INTO your_table (column1, column2) VALUES (%s, %s)"
    data_to_insert = ("value1", "value2")
    cursor.execute(insert_query, data_to_insert)
    connection.commit()  # Commit the transaction to save the changes

    # Example 3: Update data in a table
    update_query = "UPDATE your_table SET column1 = %s WHERE column2 = %s"
    data_to_update = ("new_value", "value_to_match")
    cursor.execute(update_query, data_to_update)
    connection.commit()

    # Example 4: Delete data from a table
    delete_query = "DELETE FROM your_table WHERE column1 = %s"
    data_to_delete = ("value_to_delete",)
    cursor.execute(delete_query, data_to_delete)
    connection.commit()

finally:
    # Close the cursor and the database connection
    cursor.close()
    connection.close()

    
Explanation of the cursor() and execute() methods:
    
1.cursor() Method: The cursor() method creates a cursor object associated with the database connection. 
A cursor is like a pointer or handle that allows you to execute SQL queries and fetch results from the database.
A cursor is like a pointer or handle that allows you to execute SQL queries and fetch results from the database.

2.execute() Method: The execute() method is used to execute SQL queries or commands through the cursor.
We pass an SQL query as a string to the execute() method. Optionally, we can also provide parameters to 
the query using placeholders (e.g., %s) and a tuple of values.
After executing the query, you can fetch results (in the case of SELECT queries) or commit the transaction 
(in the case of INSERT, UPDATE, DELETE) using fetchall() and commit() methods, respectively.
```

```python
Q7. Give the order of execution of SQL clauses in an SQL query.
Ans :
In an SQL query, the clauses are typically executed in a specific order, and understanding this order is 
important for constructing valid and efficient SQL statements. The general order of execution for SQL 
clauses in a SELECT statement is as follows:
    
1.FROM: The FROM clause specifies the table or tables from which data should be retrieved. It identifies 
the source tables for the query.

2.WHERE: The WHERE clause filters the rows from the tables specified in the FROM clause. It is used 
to apply conditions to the data to determine which rows should be included in the result set. Rows 
that do not meet the specified conditions are excluded.

3.GROUP BY: The GROUP BY clause is used for grouping rows with similar values in specified columns. 
This clause is typically used with aggregate functions (e.g., SUM, COUNT, AVG) to perform 
calculations on groups of data.

4.HAVING: The HAVING clause is used to filter the groups created by the GROUP BY clause based on conditions. 
It acts as a filter for grouped data, similar to the WHERE clause for individual rows.

5.SELECT: The SELECT clause determines which columns from the tables should be included in the result set. 
It specifies the data to be retrieved and presented in the output.

6.DISTINCT: The DISTINCT keyword, if used, eliminates duplicate rows from the result set based on the 
columns specified in the SELECT clause.

7.ORDER BY: The ORDER BY clause is used to sort the result set based on one or more columns. It 
specifies the order in which the rows should be displayed in the output.

8.LIMIT/OFFSET (optional): The LIMIT clause (or LIMIT/OFFSET combination) is used to restrict the
number of rows returned in the result set, while OFFSET can be used to skip a certain number of rows. 
These clauses are often used for pagination or limiting the result set size.
    
9.UNION/INTERSECT/EXCEPT (optional): If your query includes set operations like UNION, INTERSECT, 
or EXCEPT, these are applied after the individual SELECT statements are executed.
```
